## Introduction

Currently, we are working with a dataset that contains a column of text called **Tweets**.

As you probably know, *tweet* is a coined term from the popular social media site Twitter. A tweet is a status update where users post messages — from something funny to something political — on their profile, where other users can retweet the post on their own profile or comment back.

Let's load the data into our Jupyter Notebook and take a look at the first 5 rows to get a glimpse of what our columns look like.

```python
import pandas as pd
```

```python
textMining = pd.read_csv('trainTweets.csv')
textMining.head()
```

## Feature 1 — Word Count

Now we'll add a column `word_count` by applying a function that counts how many words are in the `tweet` column.

```python
# Adding a column 'word_count' and applying a
# function that counts how many words are in the column 'tweet'
textMining['word_count'] = textMining['tweet'].apply(lambda x: len(str(x).split(" ")))

# print out the first 5 rows
textMining[['tweet','word_count']].head()
```

**Expected output:**

| | tweet | word_count |
|---|---|---|
| 0 | factsguide: society now #motivation | 9 |
| 1 | retweet if you agree! | 5 |
| 2 | the next school year is the year for exams.ð... | 23 |
| 3 | we won!!! love the land!!! #allin #cavs #champ... | 13 |
| 4 | @user @user welcome here ! i'm it's so #gr... | 15 |

## Feature 2 — Character Count

We can even count how many characters are used as well.

```python
textMining['char_count'] = textMining['tweet'].str.len()  ## this also includes spaces
textMining[['tweet','char_count']].head()
```

**Expected output:**

| | tweet | char_count |
|---|---|---|
| 0 | Factsguide: Society now #motivation | 40 |
| 1 | Retweet if you agree! | 22 |
| 2 | the next school year is the year for exams.ð... | 143 |
| 3 | we won!!! love the land!!! #allin #cavs #champ... | 87 |
| 4 | @user @user welcome here ! i'm it's so #gr... | 50 |

## Feature 3 — Hashtag Count

Now, get the number of special characters used — specifically the number of hashtag (`#`) symbols.

```python
# Getting the number of the hashtag(#) symbols used
textMining['hastags'] = textMining['tweet'].apply(lambda x: len([x for x in x.split() if x.startswith('#')]))
textMining[['tweet','hastags']].head()
```

**Expected output:**

| | tweet | hastags |
|---|---|---|
| 0 | factsguide: society now #motivation | 1 |
| 1 | retweet if you agree! | 0 |
| 2 | the next school year is the year for exams.ð... | 7 |
| 3 | we won!!! love the land!!! #allin #cavs #champ... | 5 |
| 4 | @user @user welcome here ! i'm it's so #gr... | 1 |

## Conclusion

Well done! You performed feature extraction functions with Python!

In this assignment we extracted three text features from the `tweet` column:
- **`word_count`** — number of words per tweet
- **`char_count`** — number of characters per tweet (including spaces)
- **`hastags`** — number of hashtag tokens per tweet
