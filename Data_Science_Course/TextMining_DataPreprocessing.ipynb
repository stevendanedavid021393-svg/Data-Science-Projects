## Introduction

Typically, data preprocessing would be done **before** any sort of feature extraction.

In this assignment we'll cover two essential preprocessing steps for text data:
1. Converting all text to lowercase so the same word isn't counted twice (e.g. `"Case"` and `"case"` treated as one).
2. Removing punctuation to reduce noise in the dataset.

````python
import pandas as pd

textMining = pd.read_csv('trainTweets.csv')
textMining.head()
````

## Step 1 — Convert Text to Lowercase

One of the first things we'll want to do in text mining preprocessing is convert all the text into lowercase. This way one word isn't counted twice — for example, `"case"` and `"Case"` could otherwise be counted as two different words.

````python
textMining['tweet'] = textMining['tweet'].apply(lambda x: " ".join(x.lower() for x in x.split()))
textMining['tweet'].head()
````

**Expected output** (notice the top two rows now have lowercase letters):

```

1                                retweet if you agree!
2    the next school year is the year for exams.ð¯...
3    we won!!! love the land!!! #allin #cavs #champ...
4         @user @user welcome here ! i'm it's so #gr8 !
Name: tweet, dtype: object
```

## Step 2 — Remove Punctuation

We can remove punctuation as it may take up too much space in our dataset and adds little analytical value.

````python
textMining['tweet'] = textMining['tweet'].str.replace('[^\w\s]','')
textMining['tweet'].head()
````

**Expected output** (punctuation has been stripped from all tweets):

```

```

> **Note:** The three periods at the end of lines 2 and 3 are just indications that there is more text that isn't shown.

## Conclusion

Well done! You performed data preprocessing!

In this assignment we applied two key preprocessing steps to the `tweet` column:
- **Lowercasing** — ensures consistent word representation across the dataset
- **Punctuation removal** — reduces noise by stripping non-alphanumeric characters
