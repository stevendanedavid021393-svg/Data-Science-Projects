# Random Seed
**Data Science Course — Seed (random.seed) Assignment**

A **seed** is a value used to generate random data. Setting the same seed always produces the same sequence of random numbers — useful for reproducibility.

```python
import random
```

```python
#pick any seed number I chose 5
random.seed(5)
```

```python
#print a random number between 1 and 500
print(random.randint(1,500))
```

```python
#print again and get a random number
print(random.randint(1,500))
```

```python
#if you want to print the same number again
# use the same seed number from earlier
random.seed(5)
print(random.randint(1,500))
```

```python
for i in range(6):
    random.seed(5)
    print(random.randint(1,200))
    random.seed(3)
    print(random.randint(1,300))
```
