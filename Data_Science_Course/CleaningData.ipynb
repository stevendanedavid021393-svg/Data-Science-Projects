This notebook walks through common data cleaning steps on the Real Estate dataset:
normalizing a column, renaming/dropping columns, removing duplicates, and filling missing values.

```python
import pandas as pd

# Load the Real Estate CSV into a DataFrame
data = pd.read_csv('Realestate.csv')
```

```python
# --- NORMALIZE houseAge using Min-Max Scaling ---
# Formula: (value - min) / (max - min)
# This scales all values to a 0-1 range so they are comparable
min_age = data['houseAge'].min()   # 0.0
max_age = data['houseAge'].max()   # 43.8

data['houseAgeStandardized'] = (data['houseAge'] - min_age) / (max_age - min_age)
```

```python
# --- DROP the numberOfConvenienceStores column ---
# inplace=True makes the change persist directly on the DataFrame
data.drop(columns=['numberOfConvenienceStores'], inplace=True)
```

```python
# --- RENAME the transaction column to transactionDate ---
# Clearer column names improve readability and reduce confusion
data.rename(columns={'transaction': 'transactionDate'}, inplace=True)
```

```python
# --- USE .loc[] TO DISPLAY ROWS 0 THROUGH 10 ---
# .loc[] is label-based — both endpoints are INCLUSIVE
# This returns 11 rows (0, 1, 2 ... 10)
data.loc[0:10]
```

```python
# --- USE .iloc[] TO DISPLAY THE FIRST 10 ROWS ---
# .iloc[] is position-based — end index is EXCLUSIVE
# iloc[0:10] returns rows at positions 0-9 (10 rows total)
data.iloc[0:10]
```

```python
# --- FIND AND REMOVE DUPLICATE ROWS ---
# duplicated() returns True for every row that is an exact copy of a previous row
# drop_duplicates() removes those rows; inplace=True applies the change to the dataset
print("Duplicate rows found:", data.duplicated().sum())
print("Rows before removing duplicates:", len(data))

data.drop_duplicates(inplace=True)

print("Rows after removing duplicates: ", len(data))
```

```python
# --- FIND MISSING VALUES AND FILL WITH THE MEAN ---
# isnull().sum() counts NaN values in each column
print("Missing values per column:")
print(data.isnull().sum().to_string())

# Fill missing numeric values with the column mean so no data is lost
print("\nFilling missing housePriceOfUnitArea values with the column mean...")
data['housePriceOfUnitArea'] = data['housePriceOfUnitArea'].fillna(data['housePriceOfUnitArea'].mean())

print("Missing values remaining after fill:", data.isnull().sum().sum())
```
